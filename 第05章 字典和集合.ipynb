`````python
stock_day1 = [('', 0.0124), ('', 0.0111), ('', 0.0036), 
              ('', 0.0033), ('', 0.0000), (' ', -0.0071), 
              ('', -0.0078), ('', -0.0080), ('', -0.0086)]
stock_day2 = [('', 0.0231), ('', 0.0102), ('', 0.0020), 
              ('', 0.0020), ('', 0.0000), ('', -0.0065), 
              ('', -0.0102), ('', -0.0203), ('', -0.0213)]

day1_rise = {name for name, pct in stock_day1 if pct> 0}
day1_fall = {name for name, pct in stock_day1 if pct< 0}
day2_rise = {name for name, pct in stock_day2 if pct> 0}
day2_fall = {name for name, pct in stock_day2 if pct< 0}

print(' 1 ')
for name in day1_rise | day2_rise:
    print(name, end=' ')

print('\n\n 1  2 ')
for name in day1_fall & day2_rise:
    print(name, end=' ')
    
print('\n\n 1  2 ')
for name in day1_fall - day2_fall:
    print(name, end=' ')

print('\n\n 2 ')
for name in day1_rise ^ day2_rise:
    print(name, end=' ')
`````

`````python
exp_dic = {}
print('')
print('-' * 35)
i = 1
while True:
    tmp_str = input('{}'.format(i))
    if len(tmp_str) == 0:
        break
    name, amount = tmp_str.split(',')
    exp_dic.setdefault(name, 0)
    exp_dic[name] = exp_dic[name] + float(amount)
    i += 1
    
total_amount = sum(exp_dic.values())
print('-' * 35)
print('{:.2f}'.format(total_amount))
print('=' * 65)
search_name = input('')
print('-' * 35)
if search_name in exp_dic:
    print('{}{:.2f}{:.2%}'.format(search_name,
                                               exp_dic[search_name], 
                                               exp_dic[search_name] / total_amount))
else:
    print('{}'.format(search_name))
`````
