## 2.3.4 <font color=#A93226>案例：</font>股票价格增长率与涨跌幅的计算

<font size=3 face="楷体-简">增长率（growth rate）也称增长速度， 是指一定时期内某一数据指标的增长量与基期 数据的比值。 例如， 招商银行在 2002 年 4 月 9 日上市当日的收盘价为 2.59 元， 至 2020 年 12 月 31 日收盘价为 44.35 元，其收盘价折线图如图所示。</font>


<font size=3 face="楷体-简">在计算招商银行上市以来至 2020 年 12 月月底的收盘价增长率时， 可以将 2002 年 4 月 9 日的收盘价 2.59 元作为该时期期初值， 将 2020 年 12 月 31 日收盘价 44.35 元作 为该时期期末值， 用期末值除以期初值再减去 1， 得到招商银行股票价格在该段时期内的 增长率：</font>

````python
stock_beginning_price = 2.59
stock_ending_price = 44.35
(stock_ending_price / stock_beginning_price) - 1
````

<font size=3 face="楷体-简">涨跌幅是对涨跌值的描述， 用 % 表示， 涨跌幅 = 涨跌值 / 昨收盘 * 100%。 其中， 涨 跌值是当前交易日最新成交价（或收盘价）与前一交易日收盘价相减所产生的数值。 因 此，当日最新成交价比前一交易日收盘价高，涨跌幅为正，当日最新成交价比前一交易日收盘价低，则涨跌幅为负。

例如， 招商银行 2020 年 12 月 30 日收盘价为 43.05 元，2020 年 12 月 31 日收盘价为 44.35 元，所以，招商银行 2020 年 12 月 31 日的涨跌幅为：</font>

````python
price_20201230 = 43.05
price_20201231 = 44.35
(price_20201231 - price_20201230) / price_20201230
````

## 2.4.9 <font color=#A93226>案例：</font>等额本息还款法每月还款额的计算

<a href="https://www.xuetangx.com/learn/swufe08091006588/swufe08091006588/12427504/video/23498131" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>

<font size=3 face="楷体-简">等额本息还款法（等额法）， 即把按揭贷款的本金总额与利息总额相加， 平均分摊到 还款期限的每个月当中，每个月还款额是固定的，这样每个月还款额中的本金比重逐月递 增，利息比重逐月递减。

依据等额本息还款法的规则，可以推导出以下公式： 
$$ result = \frac{total * rate\_month * (1 + rate\_month)^{terms}}{(1 + rate\_month)^{terms} –1}$$

其中， result 为每月还款额， total 为贷款总额， rate_month 为银行月利率， terms 为 贷款总期数（按月算）。

编写程序，根据用户输入的贷款额度和年限，计算其利用等额本息还款法时每月的还款额。
</font>

````python
total = float(input("请输入贷款金额（元）："))
terms = int(input("请输入贷款期限（月）："))
rateYear = 0.049    #年利率为4.9%
rateMonth = rateYear / 12   #计算月利率
result = (total * rateMonth * (1 + rateMonth) ** terms) \
            / ((1 + rateMonth) ** terms - 1)
print("每月应还：",result)
````

|!=|10 != 8|True|不等于|


Python表达式支持多个比较运算。这种表达式和数学中的表达方式非常接近。

<font color=#A93226><b>【例 2-21】</b></font> <font size=3 face="楷体-简">两个数的平均数总是在这两个数的较小数和较大数之间。</font>

Python 代码如下所示：

````python
x = 10
y = 8
min(x, y) <= (x + y) / 2 <= max(x, y) 
````

字符串也可以比较大小，两个字符串的大小比较规则是：从第一个字符开始，如果第一个字符相同则顺次向后比较，直到出现不同的字符为止，以第一个不同字符的Unicode值确定大小：

````python
"abcd" < "abdd"
````

模块是第三方专门为了解决某些特定问题而编写的工具。Python本身自带了一些常用的模块，例如，math模块中具有较为复杂的求解正弦、余弦和平方根等运算，这些模块不需要安装，但是在使用前需要导入。

导入模块使用**import**关键字，Python中导入模块有不同的方法：

### （1）利用``import 模块名``
导入整个模块：

````python
import math
math.sqrt(4)
````

### （2）利用``import模块名 as 名称缩写``
导入整个模块的同时给该模块取个较短的别名：

````python
import numpy as np
np.sqrt(4)
````

### （3） 利用``import模块名.子模块名 as 名称缩写``
导入某个模块子模块的同时给该子模块取个较短的别名：

````python
import matplotlib.pyplot as plt
````

### （4）利用``from 模块名 import 函数``
导入模块中指定函数：

````python
from math import sqrt, exp
sqrt(4)
````

### （5）利用``from 模块名.子模块名 import 函数``
导入某个模块的子模块中指定函数：

## 【引导案例解析】
根据对引导案例的分析，首先需要输入姓名、日常生活支出、学习用品支出和其他方面支出等基本信息；接着计算出该月总支出；最后打印出各类支出的占比详情。

程序代码如下：

````python
stu_name = input('请输入姓名：')
living_expenses = float(input('请输入日常生活支出（元）：'))
school_things = float(input('请输入学习用品支出（元）：'))
other_expendses = float(input('其他方面支出（元）：'))
total_expenses = living_expenses + school_things + other_expendses
print('=' * 35)
print('以下是{}的支出情况：'.format(stu_name))
print('日常生活支出：{0:.2f}元，占比：{1:.2%}'.format(living_expenses, 
                                          living_expenses / total_expenses))
print('学习用品支出：{0:.2f}元，占比：{1:.2%}'.format(school_things, 
                                          school_things / total_expenses))
print('其他方面支出：{0:.2f}元，占比：{1:.2%}'.format(other_expendses, 
                                          other_expendses / total_expenses))
````
